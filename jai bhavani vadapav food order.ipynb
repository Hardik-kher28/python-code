```python
import pandas as pd
```

```python
foodlist=[]
price=[]
quantity=[]
total=[]



class jay_bhavani_shop():
   
    
    def __init__(self):
            m=[['vadapav',20] , ['puff',25] , ['pizza',80] , ['sendwich',50] , ['french fry',50] , ['dabeli',20]]
            self.menu= pd.DataFrame(m,columns=['food','price'])
            print(self.menu)
           
            self.xyz=0
            self.procede()
            
    def procede(self):
        
        ent=input('''
        Hello! welcome to our store
        1. Enter 1 for placing order
        2. Enter 2 for cancel order
        3. Enter 3 order detail
        ''')
        
        if ent == '1':
            self.order()
        elif ent == '2':
            self.cancle()
        elif ent == '3':
            self.bill()    
        else:
            print('invalid input')
            self.procede()
            
    def order(self):
        try:
        
            x=int(input('''
            1. press 1 for vadapav rs= 20
            2. press 2 for puff rs=25
            3. press 3 for pizza rs=80
            4. press 4 for sendwich rs=50
            5. press 5 for french fry rs=50
            6. press 6 for dabeli rs=20
            '''))
            if (x>0 and x<7):
                self.food=self.menu['food'][x-1]
                self.amount= self.menu['price'][x-1]
                qty=int(input('Input quantity'))
                s= self.amount*qty
                print('your order has been accepted')
                total.append(s)
                foodlist.append(self.food)
                price.append(self.amount)
                quantity.append(qty)
                self.xyz={'food':foodlist,'price':price,'quentity':qty, 'total':total}

                self.procede()

            else:
                print('invalid input')
                self.order()    
                
        except:
            print('please enter number')
            self.order()
        
    def cancle(self):
        self.xyz=0
        
        foodlist.clear()
        price.clear()
        quantity.clear()
        total.clear()
        print('your order has been cancled')
        self.procede()
          
    def bill(self):
        if  self.xyz==0:
            print('no order has been placed')
            self.procede()
        else:
            data=pd.DataFrame(self.xyz)
            print(data)
            final=data['total'].sum()+data['total'].sum()*0.04
            print('your total bill with 4% gst is:',final)
        
        
        
        
hardik= jay_bhavani_shop()    
hardik
        
        
        
```
